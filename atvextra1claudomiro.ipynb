```python
#Atividade extra claudomiro: conversão de número para binário
#Slide 2,3,4

#Esse método foca em separar o primeiro bit (a_0) colocando o número 2 em evidência.

n = 11   #Valor decimal pra converter, slide usa 13, mais na verdade, pela conversão dos binários, era pra ser (11).
a_0 = n % 2   #O operador % encontra o resto da divisão, que corresponde ao bit binário.
b_1 = n // 2   #O operador // encontra o quociente inteiro, necessário para a próxima etapa
print(f"n = 2 * b_1 + a_0")
print(f"{n} = 2 * {b_1} + {a_0} => bit a_0 = {a_0}") # o bit no caso é 0 ou 1.

#Agora a gente decompõe o quociente b_1 para encontrar a_1.

b_1 = 5  # Resultado da célula anterior (11 // 2 = 5).
a_1 = b_1 % 2  #O bit a_1 é o resto da divisão de b_1 por 2.
b_2 = b_1 // 2 #b_2 é o novo quociente para continuar o algoritmo.

print(f"b_1 = 2 * b_2 + a_1")
print(f"{b_1} = 2 * {b_2} + {a_1} => bit a_1 = {a_1}")

#Agora, encontrando os bits finais a_2 e a_3.
#b_2 já foi calculado na célula anterior como 2.

b_2 = 2    #Resultado da célula anterior (5 // 2 = 2).
a_2 = b_2 % 2   #O bit a_2 é o resto da divisão de b_2 por 2.
b_3 = b_2 // 2   # O quociente final b_3
a_3 = b_3    #Agora, o quociente b_3 torna-se o próprio bit final (a_3)

print(f"b_2 = 2 * b_3 + a_2 => bit a_2 = {a_2}")
print(f"b_3 = a_3 => bit a_3 = {a_3}")

#Lemos os bits da esquerda para a direita na escrita final, por isso começamos pelo a3, dps a2, e por fim a1.

print(f"Binário final (a3 a2 a1 a0): {a_3}{a_2}{a_1}{a_0}")
```

```python
#Atividade extra claudomiro: conversão de número para binário
#Slide 6
#Metodo de simulação da tabela de divisões para o número 347 (exemplo do slide 6), utilizando listas e for i in range.
n = 347
passos = []

#Realizando 9 divisões conforme mostrado na tabela do slide.
for k in range(9):
    resto = n % 2  #Vê o resto do número que será divido por 2, que no caso o resto é sempre 1.
    quociente = n // 2 #Vê a divisão inteira do número por 2.
    passos.append((n, resto)) #Salva o valor atual do número e o bit calculado em uma tupla dentro da lista.
    n = quociente  #Atualiza a variável n com o valor do quociente. Isso prepara o programa para a próxima divisão.

#Exibindo os resultados como na tabela.
for i, (val, bit) in enumerate(passos): #Percorre a lista de resultados guardados, fornecendo um índice i (que representa o k do slide) e os valores de N e do bit.
    print(f"N_{i} = {val} => a_{i} = {bit}") #Imprime cada linha dos valores, divididos por 2, seus novos valores, e o número em binário (0 ou 1).
```

```python
#Atividade extra claudomiro: conversão de número para binário
#Slide 7
#Primeiro, coloque uma variável em que o usuario escreve o número pra ser convertido em binário (De acordo com o slide)

 #Primeiro, definimos o número decimal e inicializar o contador (k)

n = int(input("Insira Um número para conversão: "))
k = 0
print(f"Convertendo o número {n} para binário:\n")
while True:  #Se usa while para conferir a estrutura que vai sempre se "repetir"
 #Agora, obtemos o quociente (q_k) e o resto (r_k), de acordo com o slide.

    resto_k = n % 2    #Significa o resto do número dividido por 2 (no caso vai ser sempre 1).
    quociente_k = n // 2    #Significa o valor inteiro do número dividido por 2.

#Atribuimos o bit (bit no caso é 0 ou 1 na conversão binária) a_k (que é o próprio resto(1) )
    a_k = resto_k
    print(f"Contador k={k}: N={n}, Número em Binário{k}={a_k}, Novo quociente={quociente_k}")

    if quociente_k == 0: #Nesse caso, quando o quociente(novo numéro quando dividimos por 2) for 0, a estrutura para.
        break
#Caso contrário, o novo número recebe o valor do quociente e incrementamos k (contador)
    n = quociente_k
    k = k + 1

print("\nConversão concluída!")
    #No binário sempre lemos da direita para esquerda, ou de baixo pra cima no caso. (bit mais significativo pro menos significativo)
```
